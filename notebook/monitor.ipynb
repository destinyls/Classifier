```python
import os
import json
import numpy as np

import matplotlib.pyplot as plt

def show_data(data_path):
    train_acc_path = os.path.join(data_path, "train_acc.json")
    val_acc_path = os.path.join(data_path, "val_acc.json")
    train_loss_path = os.path.join(data_path, "train_loss.json")
    val_loss_path = os.path.join(data_path, "val_loss.json")
    with open(train_acc_path) as f:
        train_acc = json.load(f)
    with open(val_acc_path) as f:
        val_acc = json.load(f)
    with open(train_loss_path) as f:
        train_loss = json.load(f)
    with open(val_loss_path) as f:
        val_loss = json.load(f)
    print("Epoch: ", len(train_acc))
    print("Best Train Acc: ", np.max(train_acc))
    print("Best Val Acc: ", np.max(val_acc))

    x = range(len(train_acc))
    plt.figure(figsize=(20,5))
    ax1 = plt.subplot(1, 2, 1)
    ax1.plot(x, train_acc,"bo-",linewidth=1,markersize=6,label="train")
    ax1.plot(x, val_acc,"go-",linewidth=1,markersize=6,label="val")
    ax1.set_title("ACC")
    ax2 = plt.subplot(1, 2, 2)
    ax2.plot(x, train_loss,"bo-",linewidth=1,markersize=6,label="train")
    ax2.plot(x, val_loss,"go-",linewidth=1,markersize=6,label="val")
    ax2.set_title("LOSS")
    
```
